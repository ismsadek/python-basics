```python
import random

def chosing_random_word():
    
    with open('sowpods.txt', 'r') as f:
        lines = f.readlines()
        
    word = random.choice(lines).strip()
    return word

def asking_letters():
    letter = input('Guess your letter: ')
    return letter.strip().upper()

def creating_the_string(word,letters_guessed):
    output = []
    for letter in word:
        if letter in letters_guessed:
            output.append(letter.upper())
        else:
            output.append("_")
    return " ".join(output)


if __name__ == '__main__':
    word_chosen = chosing_random_word()
    
    word_to_guess = set(word_chosen)
    correct_letters_guessed = set()
    incorrect_letters_guessed = set()
    num_guesses = 0
    
    
    print("Welcome ladys & gentlemens to the Hangman games!!")
    
    while(len(word_to_guess) > 0) and num_guesses < 6:
        guess = asking_letters()
        
        # check if the letter was guessed
        
        if guess in correct_letters_guessed or  guess in incorrect_letters_guessed:
            print("You already guessed that letter, say another")
            continue
        
        # if the guess is correct, we call 2 functions
        
        if guess in word_to_guess:
                word_to_guess.remove(guess)
                correct_letters_guessed.add(guess)
            
        else:
            incorrect_letters_guessed.add(guess)
            num_guesses += 1
        
        
        word = creating_the_string(word_chosen,correct_letters_guessed)
        
        print(word)
        print("Hey you have {} tries left".format(6-num_guesses))
        
    if num_guesses < 6:
        print("Yeaahhh!! You have guessed the word {}".format(word_chosen))
    else:
        print("Sorry, game over! The word was {}".format(word_chosen))
        
    

    
```
